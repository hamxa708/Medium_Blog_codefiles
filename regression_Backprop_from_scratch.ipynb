```python
def initialize_parameters(layer_dims):
  #so that results are reproducable
  np.random.seed(3)
  #to store the initilaized parameters
  parameters = {}
  L = len(layer_dims)
  #iterates over layers of neural network. Input is not neuron layer; 1 is hidden layer here.
  for l in range(1, L):
  # initizaling weights and biases
    parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
    parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))


  return parameters
```

```python
 initialize_parameters([2,2,1])
 # The dictionary stores all weights of particular layers in a matrix. Like 1st layer as 4 weights and 2nd has 2.
 # the biases are stored in a column vector.
 # we have taken 2 layers here.
```

```python
def linear_forward(O_previous,w,b):
# O_previous refers to output from previous neuron
  Z=np.dot(w.T,O_previous)+b

  return Z
```

```python
# Forward Propagation
def L_layer_forward(X, parameters):         # X is an example

  A = X
  L = len(parameters) // 2                  # number of layers

  for l in range(1, L+1):

    O_previous = A
    Wl = parameters['W' + str(l)]
    bl = parameters['b' + str(l)]
    #print("A"+str(l-1)+": ", O_previous)
    #print("W"+str(l)+": ", Wl)
    #print("b"+str(l)+": ", bl)
    #print("--"*20)

    A = linear_forward(O_previous, Wl, bl)
    #print("A"+str(l)+": ", A)
    #print("**"*20)


  return A,O_previous
```

```python
def update_parameters(parameters,y,y_hat,A1,X):
  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

```python
parameters = initialize_parameters([2, 2, 1])
epochs = 10  # Set the total number of epochs

for i in range(epochs):
    Loss = []  # List to collect the loss for the current epoch

    for j in range(df.shape[0]):
        # Extract input features and the corresponding label
        X = df[['hours_studied', 'previous_score']].values[j].reshape(2, 1)
        y = df[['exam_score']].values[j][0]

        # Perform forward pass and compute the output
        y_hat, A1 = L_layer_forward(X, parameters)
        y_hat = y_hat[0][0]  # Extract scalar value from y_hat

        # Update parameters based on the error
        update_parameters(parameters, y, y_hat, A1, X)

        # Collect the loss and store the last prediction
        Loss.append((y - y_hat) ** 2)

    # Print epoch and loss at multiples of 2 epochs
    if (i % 2) == 0:
        print('Epoch -', i + 1, 'Loss -', np.array(Loss).mean())
```
